MM Kinetics, adapted from coursework

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
```

```python
# Mutant enzyme H418N 30mM TES buffer 0.01mM Mg2+, 150 mM Na+, pH 7.0, temp 25C
K_cat = 5.5
K_M = 0.05
# Variables
E_0 = 0.1
# ODE
def MM_dynamics(t,y,K_cat,K_M,E_0):
    # y[0] = B product
    # y[1] = A reactant
    if y[1] <= 0:
        y[1] = 0
    dydt = np.zeros(2)
    v = K_cat*E_0*(y[1]/(K_M+y[1]))
    dydt[0] = v
    dydt[1] = -v
    return dydt
# Initial Conditions
y0 = [0,100]
# time span
t = np.linspace(0,220,100)
tspan = [t[0],t[-1]]

ode_sol = solve_ivp(lambda t,y:MM_dynamics(t,y,K_cat,K_M,E_0),tspan,y0,t_eval=t)
plt.plot(t,ode_sol.y[0],'k-')
plt.plot(t,ode_sol.y[1],'r--')
plt.xlabel('time [s]')
plt.ylabel('concentration [mM]')
plt.legend(['product','reactant']);
```

Gene Regulation

```python
##concentration of protein no inhibitor
## formulas adapted from Munro et al. 2016, 'Aspects of protein-DNA interactions: a review of quantitative thermodynamics theory for modelling synthetic circuits utilizing LacI and CI repressors, IPTG and the reporter gene LacZ

def gene_expression(t,y,k_in,k_deg,k_vol,P_prom):
    dydt = (k_in * P_prom * A_bgal) - (k_vol + k_deg) * (prot_total)
    return dydt

A_bgal = 30 ##beta galactose proteins produced per mRNA molecule

k_in = 0.09 ##rate production

k_deg = 0.0000001 ##rate constant protein degradation, very small

k_vol = 0.00029 ##E. coli expanding at constant rate volume, doubling time 40 min

P_prom = 3 ##concentration of promoter bound by RNA polymerase 

prot_total = (k_in / (k_vol + k_deg)) * (P_prom) * (A_bgal)

y = [0,100]
t = np.linspace(0,10,100)
tspan = [t[0],t[-1]]

ode_sol = solve_ivp(lambda t,y:gene_expression(t,y,k_in,k_deg,k_vol,P_prom),tspan,y,t_eval=t)
plt.plot(t,ode_sol.y[0],'k-')
plt.xlabel('time [s]')
plt.ylabel('protein concentration')
```
